```python
import pandas as pd
import numpy as np
from scipy.interpolate import griddata
from scipy.interpolate.interpolate import interp1d
import plotly.graph_objects as go
```

```python
df1 = pd.read_csv('data/1408_cz.csv')
df2 = pd.read_csv('data/1410_cz.csv')
df3 = pd.read_csv('data/1412_cz.csv')
df1['thickness'] = 8
df2['thickness'] = 10
df3['thickness'] = 12
```

```python
def layout():
   camera = dict(eye=dict(x=0.1, y=-2, z=0.1))
   scene = dict(
      xaxis = dict(title='Alpha'),
      yaxis = dict(title='Thickness'),
      zaxis = dict(title='Cz')
   )
   legend = dict(font = {'size': 12}, itemsizing = 'trace')
   return go.Layout(margin=dict(l=0, r=0, b=0, t=0), 
                     width=1024, height=768, 
                     scene=scene,
                     scene_camera=camera,
                     legend_title_text='Data',
                     legend=legend)

def plot2d():
        fig = go.Figure(layout=go.Layout(xaxis = dict(title='Alpha'), yaxis = dict(title='Cz')))
        fig.add_trace(go.Scatter(x=df1.x, 
                                   y=df1['cz'],
                                   mode='lines',
                                   name='1408'))
        fig.add_trace(go.Scatter(x=df2.x, 
                                   y=df2['3'],
                                   mode='lines',
                                   name='1410'))
        fig.add_trace(go.Scatter(x=df3.x, 
                                   y=df3['3'],
                                   mode='lines',
                                   name='1412'))
        fig.show()

def plot3d():
        fig = go.Figure(layout=layout())
        fig.add_trace(go.Scatter3d(x=df1.x, 
                                   y=df1.thickness,
                                   z=df1['cz'],
                                   mode='lines',
                                   name='1408'))
        fig.add_trace(go.Scatter3d(x=df2.x, 
                                   y=df2.thickness,
                                   z=df2['3'], 
                                   mode='lines',
                                   name='1410'))
        fig.add_trace(go.Scatter3d(x=df3.x, 
                                   y=df3.thickness,
                                   z=df3['3'], 
                                   mode='lines',
                                   name='1412'))
        fig.show()
```

```python
plot2d()
```

```python
plot3d()
```
